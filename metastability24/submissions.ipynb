```python
def endtime(data, i):
    if data["Type"][i] == 'Keynote':
        return (pd.to_datetime(data.Time[i], format='%H%M') + pd.Timedelta(minutes=40)).strftime('%H%M')
    else:
        return (pd.to_datetime(data.Time[i], format='%H%M') + pd.Timedelta(minutes=20)).strftime('%H%M')
    
days = ['Tuesday', 'Wednesday', 'Thursday']
```

```python
for i in range(len(data)):
    for j in [1]:
        if data["Type"][i] == 'Keynote' and data['Day'][i]==j:
            print("<h4>{}</h4>".format(str(data["Title"][i])))
            if type(data["Coauthors"][i]) == float:
                print("<p><b>{} {}</b>, {}<br><i>{}, {}-{}, Room {}</i></p>".format(
                str(data['Firstname'][i]), str(data["Lastname"][i]), str(data["Affiliation"][i]),
                days[data['Day'][i]-1], pd.to_datetime(data['Time'][i], format='%H%M').strftime(format='%H:%M'), pd.to_datetime(endtime(data, i), format='%H%M').strftime(format='%H:%M'), data['Track'][i]
            ))
            else:
                print("<p><b>{} {}</b>, {}<br>with {}<br><i>{}, {}-{}, Room {}</i></p>".format(
                str(data['Firstname'][i]), str(data["Lastname"][i]), str(data["Affiliation"][i]), str(data["Coauthors"][i]),
                days[data['Day'][i]-1], pd.to_datetime(data['Time'][i], format='%H%M').strftime(format='%H:%M'), pd.to_datetime(endtime(data, i), format='%H%M').strftime(format='%H:%M'), data['Track'][i]
            ))
            print("<p>{}</p>".format(str(data["Abstract"][i])))
            print("")
```

```python
def print_schedule(data, day):
    print(r'<h3 id="schedule-heading">{}, {} September</h3>'.format(days[day-1], day+2))
    print(r'<div class="schedule" aria-labelledby="schedule-heading">')
    print(r'<span class="track-slot" aria-hidden="true" style="grid-column: track-1; grid-row: tracks;">Room 1</span>')
    print(r'<span class="track-slot" aria-hidden="true" style="grid-column: track-2; grid-row: tracks;">Room 2</span>')
    print(r'<h2 class="time-slot" style="grid-row: time-0900;">9:00am</h2>')
    print("")
    for i in range(len(data)):
        if data.Day[i] == day:
            day = data['Day'][i]
            if data['Track'][i] == 3:
                track = "track-1-start / track-2-end"
            else:
                track = "track-{}".format(data['Track'][i])
            start = data['Time'][i]
            end = endtime(data, i)
            print(r'<div class="session session-{} track-{}" style="grid-column: {}; grid-row: time-{} / time-{};">'.format(
                day, data['Track'][i], track, start, end))
            print(r'<h3 class="session-title"><a href="#">{}</a></h3>'.format(
                data.Title[i]
            ))
            print(r'<span class="session-presenter">{} {}</span>'.format(
                data.Firstname[i], data.Lastname[i]
            ))
            print(r'<span class="session-time">{} - {}</span>'.format(
                pd.to_datetime(start, format='%H%M').strftime(format='%H:%M'), pd.to_datetime(end, format='%H%M').strftime(format='%H:%M')
            ))
            print(r'<span class="session-track">Room {}</span>'.format(
                2-data.Track[i]%2
            ))
            print("</div>")
    print("</div>")
    print("")

print_schedule(data, 3)
#print_schedule(data, 2)
#print_schedule(data, 3)
```
